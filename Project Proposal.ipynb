# Motivation

In the modern world, we have an abundance of information. This information has been used to make large language inference models. It is becoming an increasing challenge to trust the information that is online. As models get larger and larger, the text becomes harder to trace by human intuition. Making it hard to identify if the author is a human or not. There is a growing anti-AI sentiment, and the push back is founded on the systemic issues with privacy, education, content ownership, and copyright infringement.

# Problem

the problem is a clarification problem where we wish to classify english text into 2 classes human and AI generated content. This model needs to be small enough to send over https and should be performant enough to run on a phone
